##### 1. tree construction

```python
G = nx.DiGraph()
lst = [(0,1), (0,2), (1,3), (1,4), (2,5)]
G.add_edges_from(lst)
```

---

##### 2. tree visualization

```python
fig = plt.figure(figsize=(3,2),facecolor='white')


H = nx.bfs_tree(G, source=0)

pos=graphviz_layout(H, prog='dot')

nx.draw(H,
        pos,
        node_size = 300,
        arrows=True,
        edgecolors = 'black', # node boder color
        node_color='white',
        with_labels = True,
#         alpha = .9,
       )

plt.show()
```

---

##### 3. results

```python
# the cascade virality of G is:
print(pure_cascade_virality(G))
print(recursive_cascade_virality(G))
```
